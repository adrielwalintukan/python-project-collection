```python
# Tabel
from tabulate import tabulate

# Perhitunggan nilai
def hitung_nilai_akhir(nilai_assignment, nilai_quiz, nilai_project, nilai_mid, nilai_final):
    persentase_assignment = 0.2
    nilai_assignment_persen = sum(nilai_assignment) / len(nilai_assignment) * persentase_assignment

    persentase_quiz = 0.15
    nilai_quiz_persen = sum(nilai_quiz) / len(nilai_quiz) * persentase_quiz

    persentase_project = 0.15
    nilai_project_persen = nilai_project * persentase_project

    persentase_mid = 0.25
    nilai_mid_persen = nilai_mid * persentase_mid

    persentase_final = 0.25
    nilai_final_persen = nilai_final * persentase_final

    assignment = nilai_assignment_persen
    quiz = nilai_quiz_persen
    project = nilai_project_persen
    mid = nilai_mid_persen
    final = nilai_final_persen

    nilai_akhir = assignment + quiz + project + mid + final
    return nilai_akhir, nilai_assignment_persen, nilai_quiz_persen, nilai_project_persen, nilai_mid_persen, nilai_final_persen

# Tampilkan nilai dalam tabel
def tampilkan_hasil(data_mahasiswa, assignment_headers, quiz_headers, menu_option):
    if menu_option == 1:
        print('Nilai akhir Mahasiswa')
        headers = ['NIM', 'Nama', 'Nilai Akhir']
        simplified_data = [[data[0], data[1], data[7]] for data in data_mahasiswa]
        print(tabulate(simplified_data, headers=headers, tablefmt='fancy_grid'))
    else:
        headers = ['NIM', 'Nama', 'Ass 1 | Ass 2 | Ass 3 | 20%', 'Quiz 1 | Quiz 2 | 15%', 'Project | 15%', 'Mid | 25%', 'Final | 25%', 'Nilai Akhir', 'Grade']
        print(tabulate(data_mahasiswa, headers=headers, tablefmt='fancy_grid'))

# Nilai tertinggi dan terendah
def info(data_mahasiswa):
    nilai_akhir_list = [float(data[7]) for data in data_mahasiswa]
    tertinggi_nilai_akhir = max(nilai_akhir_list)
    terendah_nilai_akhir = min(nilai_akhir_list)

    mahasiswa_tertinggi = [data[1] for data in data_mahasiswa if float(data[7]) == tertinggi_nilai_akhir]
    mahasiswa_terendah = [data[1] for data in data_mahasiswa if float(data[7]) == terendah_nilai_akhir]

    jumlah_tidak_lulus = sum(1 for data in data_mahasiswa if float(data[7]) < 67)

    print('\nInformasi:')
    print(f"Yang mendapatkan nilai akhir tertinggi adalah: {', '.join(mahasiswa_tertinggi)} dengan nilai {tertinggi_nilai_akhir}")
    print(f"Yang mendapatkan nilai akhir terendah adalah: {', '.join(mahasiswa_terendah)} dengan nilai {terendah_nilai_akhir}")
    print(f"Jumlah Mahasiswa yang Tidak Lulus: {jumlah_tidak_lulus}")

# Urutkan tertinggi -> terendah
def urutkan_nilai_tertinggi_ke_terendah(data_mahasiswa):
    print('Data Mahasiswa Setelah Diurutkan berdasarkan Nilai Akhir (Tertinggi -> Terendah):')
    tampilkan_hasil(sorted(data_mahasiswa, key=lambda x: float(x[7]), reverse=True), [], [], 1)

# Frekuensi grade
def frekuensi_grade(data_mahasiswa):
    grades = [data[8] for data in data_mahasiswa]
    grade_frekuensi = {grade: grades.count(grade) for grade in set(grades)}

    print('Frekuensi Grade:')
    for grade, frekuensi in grade_frekuensi.items():
        print(f'Grade {grade}: {frekuensi} orang')

# Tampilkan semua hasil
def tampilkan_semua_hasil(data_mahasiswa, assignment_headers, quiz_headers):
    print('\nSemua Hasil:')
    print('-' * 50)

    print('\n1. Tampilkan Nilai Akhir Mahasiswa:')
    tampilkan_hasil(data_mahasiswa, assignment_headers, quiz_headers, 1)

    print('\n2. Tampilkan Grade Mahasiswa:')
    for data in data_mahasiswa:
        print(f'{data[1]} (NIM: {data[0]}): {data[8]}')

    print('\n3. Tampilkan Tabel Lengkap dan Informasi:')
    tampilkan_hasil(data_mahasiswa, [], [], 3)
    info(data_mahasiswa)

    print('\n4. Urutkan Nilai Akhir (Tertinggi ke Terendah):')
    urutkan_nilai_tertinggi_ke_terendah(data_mahasiswa)

    print('\n5. Tampilkan Frekuensi Grade:')
    frekuensi_grade(data_mahasiswa)

# Menu
def menu():
    print(
        '\n-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-B-A-T-A-S-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-\n\n'
        '\t\tPilihan MENU\n'
        '1. Tampilkan Nilai Akhir Mahasiswa\n'
        '2. Tampilkan Grade Mahasiswa\n'
        '3. Tampilkan Tabel Lengkap dan Informasi\n'
        '4. Urutkan Nilai Akhir (Tertinggi ke Terendah)\n'
        '5. Tampilkan Frekuensi Grade\n'
        '6. Tampilkan Semua Hasil (Menu 1-5)\n'
        '7. Hentikan Program/Ulangi Program\n'
        '8. Mencetak Hasil ke Dalam File'
    )

# Mulai
data_mahasiswa = []
print('\t- PROGRAM MENGHITUNG NILAI MAHASISWA -\t')
print('\nSilahkan mengikuti petunjuk-petunjuk yang ada.')

# Input
jumlah_mahasiswa = int(input('\nMasukkan jumlah Mahasiswa yang akan dihitung nilainya: '))

for i in range(jumlah_mahasiswa):
    nama = input(f'\nMasukkan Nama Mahasiswa ke-{i + 1}: ')
    nim = input(f'Masukkan NIM: ')

    print('Silahkan masukkan 3 nilai Assignment')
    nilai_assignment = [float(input(f'\tnilai Assignment ke-{x + 1}: ')) for x in range(3)]
    print('Silahkan masukkan 2 nilai Quiz')
    nilai_quiz = [float(input(f'\tnilai Quiz ke-{y + 1}: ')) for y in range(2)]
    print('Silahkan masukkan nilai Project')
    nilai_project = float(input(f'\tnilai Project: '))
    print('Silahkan masukkan nilai Mid')
    nilai_mid = float(input(f'\tnilai Mid: '))
    print('Silahkan masukkan nilai Final')
    nilai_final = float(input(f'\tnilai Final '))

    nilai_akhir, persentase_assignment, persentase_quiz, persentase_project, persentase_mid, persentase_final = hitung_nilai_akhir(nilai_assignment, nilai_quiz, nilai_project, nilai_mid, nilai_final)

    if nilai_akhir >= 91 and nilai_akhir <= 100:
        grade = 'A'
    elif nilai_akhir >= 85:
        grade = 'A-'
    elif nilai_akhir >= 82:
        grade = 'B+'
    elif nilai_akhir >= 78:
        grade = 'B'
    elif nilai_akhir >= 75:
        grade = 'B-'
    elif nilai_akhir >= 70:
        grade = 'C+'
    elif nilai_akhir >= 67:
        grade = 'C'
    elif nilai_akhir >= 60:
        grade = 'C-'
    elif nilai_akhir >= 40:
        grade = 'D'
    else:
        grade = 'F'

    assignment_data = ' | '.join(map(str, nilai_assignment)) + f' | {persentase_assignment:.2f}'
    quiz_data = ' | '.join(map(str, nilai_quiz)) + f' | {persentase_quiz:.2f}'
    project_data = f'{nilai_project} | {persentase_project:.2f}'
    mid_data = f'{nilai_mid} | {persentase_mid:.2f}'
    final_data = f'{nilai_final} | {persentase_final:.2f}'

    data_mahasiswa.append([nim, nama, assignment_data, quiz_data, project_data, mid_data, final_data, f'{nilai_akhir:.2f}', grade])

while True:
    menu()
    pilihan = int(input('\nMasukkan pilihan anda (1/2/3/4/5/6/7/8): '))

    if pilihan == 1:
        tampilkan_hasil(data_mahasiswa, [f'Ass {i + 1}' for i in range(3)], [f'Quiz {i + 1}' for i in range(2)], 1)
    elif pilihan == 2:
        print('\nGrade Mahasiswa:')
        for data in data_mahasiswa:
            print(f'{data[1]} (NIM: {data[0]}): {data[8]}')
    elif pilihan == 3:
        tampilkan_hasil(data_mahasiswa, [], [], 3)
        info(data_mahasiswa)
    elif pilihan == 4:
        urutkan_nilai_tertinggi_ke_terendah(data_mahasiswa)
    elif pilihan == 5:
        frekuensi_grade(data_mahasiswa)
    elif pilihan == 6:
        tampilkan_semua_hasil(data_mahasiswa, [f'Ass {i + 1}' for i in range(3)], [f'Quiz {i + 1}' for i in range(2)])
    elif pilihan == 7:
        ulangi = input('\nApakah Anda ingin hentikan program/ulangi program? (berhenti/ulangi): ')
        if ulangi.lower() != 'ulangi':
            print('\nProgram Selesai.')
            print('TERIMA KASIH TELAH MENGGUNAKAN PROGRAM SAYA :)')
            break
        else:
            data_mahasiswa = []
    elif pilihan == 8:
   # Bonus
        with open('Student.txt', 'w') as file:
            file.write("-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-B-A-T-A-S-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-#-\n\n")
            file.write("\t\tHasil Akhir Program\n\n")
            headers = ['NIM', 'Nama', 'Ass 1 | Ass 2 | Ass 3 | 20%', 'Quiz 1 | Quiz 2 | 15%', 'Project | 15%', 'Mid | 25%', 'Final | 25%', 'Nilai Akhir', 'Grade']
            file.write(tabulate(data_mahasiswa, headers=headers, tablefmt='fancy_grid'))
            file.write("\n\n")
            file.write("\n\nData Mahasiswa Setelah Diurutkan berdasarkan Nilai Akhir (Tertinggi -> Terendah):\n")
            sorted_data = sorted(data_mahasiswa, key=lambda x: float(x[7]), reverse=True)
            file.write(tabulate(sorted_data, headers=headers, tablefmt='fancy_grid'))
            file.write("\n\n")

            nilai_akhir_list = [float(data[7]) for data in data_mahasiswa]
            tertinggi_nilai_akhir = max(nilai_akhir_list)
            terendah_nilai_akhir = min(nilai_akhir_list)

            mahasiswa_tertinggi = [data[1] for data in data_mahasiswa if float(data[7]) == tertinggi_nilai_akhir]
            mahasiswa_terendah = [data[1] for data in data_mahasiswa if float(data[7]) == terendah_nilai_akhir]

            jumlah_tidak_lulus = sum(1 for data in data_mahasiswa if float(data[7]) < 67)

            file.write('\nInformasi:')
            file.write(f"\nYang mendapatkan nilai akhir tertinggi adalah: {', '.join(mahasiswa_tertinggi)} dengan nilai {tertinggi_nilai_akhir}")
            file.write(f"\nYang mendapatkan nilai akhir terendah adalah: {', '.join(mahasiswa_terendah)} dengan nilai {terendah_nilai_akhir}")
            file.write(f"\nJumlah Mahasiswa yang Tidak Lulus: {jumlah_tidak_lulus}")

            file.write('\n\nFrekuensi Grade:')
            grades = [data[8] for data in data_mahasiswa]
            grade_frekuensi = {grade: grades.count(grade) for grade in set(grades)}

            for grade, frekuensi in grade_frekuensi.items():
                file.write(f'\nGrade {grade}: {frekuensi} orang')

        print("\nHasil telah dicetak dalam bentuk file ke dalam Student.txt.")
    else:
        print('Pilihan tidak valid. Silakan pilih kembali.')
```

```python
# Menu 1
def insert_data(narapidana_list):
  print('\nTambahkan data Narapidana')
  nik = input('Masukkan NIK Narapidana: ')
  nama = input('Masukkan Nama Narapidana: ')
  umur = input('Masukkan Umur Narapidana: ')
  asal = input('Masukkan Asal Narapidana: ')
  kejahatan = input('Masukkan Kejahatan yang dilakukan: ')
  lama_tahanan = input('Masukkan Lama Tahanan: ')

  narapidana = (nik, nama, umur, asal, kejahatan, lama_tahanan)
  narapidana_list.append(narapidana)
  print('Data Narapidana berhasil ditambahkan.')

# Menu 2
def delete_data(narapidana_list):
  print('\nHapus data Narapidana')
  delete = int(input('Masukkan data Narapidana ke berapa yang ingin dihapus: '))
  if delete < 1 or delete > len(narapidana_list):
    print('Data yang anda masukkan tidak valid.')
  else:
    del narapidana_list[delete-1]
    print(f'Data Narapidana ke-{delete} berhasil dihapus.')

# Menu 3
def display_data(narapidana_list):
  print('\nMenampilkan semua data Narapidana')
  print('Daftar Narapidana:')
  if not narapidana_list:
    print('Tidak ada data.')
  else:
    for i in range(len(narapidana_list)):
      napi = narapidana_list[i]
      print(f'{i+1}. - NIK: {napi[0]}\n   - NAMA: {napi[1]}\n   - UMUR: {napi[2]}\n   - ASAL: {napi[3]}\n   - Jenis Kejahatan: {napi[4]}\n   - Lama Tahanan: {napi[5]}')

# Menu 4
# Saya menggunakan algoritma Merge Sort.
def sorting_data(narapidana_list, sort_by):
    if len(narapidana_list) <= 1:
        return narapidana_list

    mid = len(narapidana_list) // 2
    left_half = narapidana_list[:mid]
    right_half = narapidana_list[mid:]

    left_half = sorting_data(left_half, sort_by)
    right_half = sorting_data(right_half, sort_by)

    return merge(left_half, right_half, sort_by)

def merge(left, right, sort_by):
    result = []
    left_index, right_index = 0, 0

    if sort_by == '1':
        while left_index < len(left) and right_index < len(right):
            if left[left_index]['nama'] < right[right_index]['nama']:
                result.append(left[left_index])
                left_index += 1
            else:
                result.append(right[right_index])
                right_index += 1
    elif sort_by == '2':
        while left_index < len(left) and right_index < len(right):
            if int(left[left_index]['umur']) < int(right[right_index]['umur']):
                result.append(left[left_index])
                left_index += 1
            else:
                result.append(right[right_index])
                right_index += 1
    elif sort_by == '3':
        while left_index < len(left) and right_index < len(right):
            if int(left[left_index]['lama_tahanan']) < int(right[right_index]['lama_tahanan']):
                result.append(left[left_index])
                left_index += 1
            else:
                result.append(right[right_index])
                right_index += 1

    while left_index < len(left):
        result.append(left[left_index])
        left_index += 1

    while right_index < len(right):
        result.append(right[right_index])
        right_index += 1

    return result

def menu4(narapidana_list):
    print('\nUrutkan data Narapidana berdasarkan:')
    print('1. Abjad nama tahanan')
    print('2. Umur tahanan')
    print('3. Lama tahanan')
    sort_by = input('Masukkan pilihan (1/2/3): ')
    narapidana_list = sorting_data(narapidana_list, sort_by)
    print('\nData telah diurutkan.')
    print('Silakan pilih menu 3 untuk menampilkan hasil.')

# Menu 5
# Alasan saya menggunakan algoritma Linear Search dalam program ini adalah karena kita hanya memiliki satu list yang berisi data narapidana.
# Algoritma Linear Search memiliki kompleksitas waktu O(n), yang berarti waktu eksekusi akan meningkat linear sesuai dengan jumlah data yang ada.
# Dalam program saya ini, karena kita hanya memiliki satu list yang berisi data narapidana, algoritma Linear Search akan menjadi pilihan yang cocok untuk
# mencari data narapidana berdasarkan NIK.
def searching_data(narapidana_list):
  print('\nCari data Narapidana')
  nik = input('Masukkan NIK narapidana yang akan dicari: ')
  found = False
  for i, narapidana in enumerate(narapidana_list):
    if narapidana[0] == nik:
      print('Data Narapidana ditemukan:')
      print(f'   - NIK: {narapidana[0]}\n   - NAMA: {narapidana[1]}\n   - UMUR: {narapidana[2]}\n   - ASAL: {narapidana[3]}\n   - Jenis Kejahatan: {narapidana[4]}\n   - Lama Tahanan: {narapidana[5]}')
      found = True
      break
  if not found:
    print('Data Narapidana tidak ditemukan.')


def menu():
  print('\nPilihan Menu:')
  print('1. Tambahkan Data Narapidana')
  print('2. Hapus Data Narapidana')
  print('3. Tampilkan Semua Data Narapidana')
  print('4. Urutkan Data Narapidana')
  print('5. Cari Data Narapidana')
  print('6. Exit/Keluar')

def main():
  narapidana_list = []
  print('\n\t--- PROGRAM INFORMASI NARAPIDANA ---')
  while True:
    menu()
    pilihan = input('Masukkan pilihan anda (1-6): ')

    if pilihan == '1':
      insert_data(narapidana_list)
    elif pilihan == '2':
      delete_data(narapidana_list)
    elif pilihan == '3':
      display_data(narapidana_list)
    elif pilihan == '4':
      menu4(narapidana_list)
    elif pilihan == '5':
      searching_data(narapidana_list)
    elif pilihan == '6':
      print('\nProgram selesai.')
      print('--- Terima Kasih ---')
      break
    else:
      print('\nPilihan menu tidak valid.')

if __name__ == "__main__":
  main()
```
