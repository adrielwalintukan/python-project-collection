# **Queue Line Management System**

```python
from collections import deque
import datetime
import pytz

class Antrian:
  def __init__(self):
    self.antrian = deque()
    self.nomor_terakhir = 0

  def enqueue(self):
    self.nomor_terakhir += 1
    self.antrian.append(self.nomor_terakhir)
    print("\nBerhasil mendapatkan nomor antrian")
    print(f"Nomor antrian Anda: {self.nomor_terakhir}")
    print("Silahkan proses pesanan Anda.")

  def dequeue(self):
    if len(self.antrian) == 0:
      print("\nAntrian kosong")
    else:
      print(f"\nNomor antrian {self.antrian.popleft()} telah diproses.")

  def cancel(self):
    batal = int(input("Masukkan nomor antrian Anda yang akan dibatalkan: "))
    if batal in self.antrian:
      self.antrian.remove(batal)
      print(f"\nNomor antrian {batal} berhasil dibatalkan")
    else:
      print("\nNomor antrian tidak ditemukan")

  def search(self):
    dicari = input("Masukkan nomor antrian: ")
    if int(dicari) in self.antrian:
      posisi = self.antrian.index(int(dicari)) + 1
      print(f"\nAntrian saat ini: {list(self.antrian)}")
      if posisi == 1:
        print(f"Anda saat ini berada di urutan ke-{posisi}. Pesanan Anda akan segera diproses.")
      else:
        print(f"Anda saat ini berada di urutan ke-{posisi}. Masih ada {posisi - 1} antrian lagi sebelum pesanan Anda diproses.")
    else:
      print("\nNomor antrian tidak ditemukan")

  def check(self):
    if self.antrian:
      wita = pytz.timezone('Asia/Makassar')
      waktu = datetime.datetime.now(wita).strftime("%H:%M")
      print(f"\nAntrian saat ini pada jam {waktu} WITA")
      print(f"Berjumlah: {len(self.antrian)} antrian")
      print(f"List antrian yang sedang diproses saat ini: {list(self.antrian)} ")
    else:
      print("\nBelum ada antrian")

def main_menu():
  print("""\n\tMENU
1. Ambil nomor antrian
2. Tampilkan status antrian yang sedang diproses saat ini
3. Selanjutnya
4. Keluar""")

def sub_menu():
  print("""\n\tMENU
1. Proses pesanan
2. Cancel antrian
3. Cek status antrian anda
4. Ambil nomor antrian
5. Kembali ke menu utama
6. Keluar""")

def antrian_menu():
  antrian = Antrian()
  print("Antrian:")
  while True:
    main_menu()
    pil = input("Silahkan pilih menu: ")
    if pil == "1":
      antrian.enqueue()
      while True:
        sub_menu()
        pil2 = input("Silahkan pilih menu: ")
        if pil2 == "1":
          antrian.dequeue()
        elif pil2 == "2":
          antrian.cancel()
        elif pil2 == "3":
          if len(antrian.antrian) > 0:
            antrian.search()
          else:
            print("\nAntrian kosong")
        elif pil2 == "4":
          antrian.enqueue()
        elif pil2 == "5":
          break
        elif pil2 == "6":
          berhenti = input("\nApakah Anda yakin ingin keluar? (ya/tidak): ")
          if berhenti.lower() == "ya":
            print("\nAnda telah keluar.")
            print("--- TERIMA KASIH ---")
            return
          elif berhenti.lower() == "tidak":
            antrian.antrian = deque()
          else:
            print("\nPilihan tidak valid. Silahkan ketik ya/tidak.")
        else:
          print("\nPilihan menu Anda tidak valid. Silahkan pilih menu yang sesuai dari 1-6")
    elif pil == "2":
        antrian.check()
    elif pil == "3":
      while True:
        sub_menu()
        pil2 = input("Silahkan pilih menu: ")
        if pil2 == "1":
          antrian.dequeue()
        elif pil2 == "2":
          antrian.cancel()
        elif pil2 == "3":
          if len(antrian.antrian) > 0:
            antrian.search()
          else:
            print("\nAntrian kosong")
        elif pil2 == "4":
          antrian.enqueue()
        elif pil2 == "5":
          break
        elif pil2 == "6":
          berhenti = input("\nApakah Anda yakin ingin keluar? (ya/tidak): ")
          if berhenti.lower() == "ya":
            print("\nAnda telah keluar.")
            print("--- TERIMA KASIH ---")
            return
          elif berhenti.lower() == "tidak":
            antrian.antrian = deque()
          else:
            print("\nPilihan tidak valid. Silahkan ketik ya/tidak.")
        else:
          print("\nPilihan menu Anda tidak valid. Silahkan pilih menu yang sesuai dari 1-6")
    elif pil == "4":
      berhenti = input("\nApakah Anda yakin ingin keluar? (ya/tidak): ")
      if berhenti.lower() == "ya":
        print("\nAnda telah keluar.")
        print("--- TERIMA KASIH ---")
        break
      elif berhenti.lower() == "tidak":
        antrian.antrian = deque()
      else:
        print("\nPilihan tidak valid. Silahkan ketik ya/tidak.")
    else:
      print("\nPilihan menu Anda tidak valid. Silahkan pilih menu yang sesuai dari 1-4")

antrian_menu()
```
