```python
# Saya menggunakan algoritma Merge Sort. Alasan: Karena algoritma Merge Sort time complexitynya adalah o(log n).
# Jadi lebih efisien dibandingkan dengan algoritma sorting yang lain.
# Dan juga alasan lain saya menggunakan algoritma Merge Sort ini adalah karena ma'am memberikan contoh menggunakan algoritma Merge Sort.

# Saya menggunakan algoritma Linear Search. Alasan: Karena kita hanya memiliki satu list yang berisi data narapidana.
# Algoritma Linear Search memiliki kompleksitas waktu O(n), yang berarti waktu eksekusi akan meningkat linear sesuai dengan jumlah data yang ada.
# Dalam program saya ini, karena kita hanya memiliki satu list yang berisi data narapidana, algoritma Linear Search akan menjadi pilihan yang cocok untuk
# mencari data narapidana berdasarkan NIK. Algoritma Linear Search juga bisa mencari data walaupun datanya belum terurut.

# Menu 1
def insert_data(narapidana_list):
  print('\nTambahkan data Narapidana')
  nik = input('Masukkan NIK Narapidana: ')
  nama = input('Masukkan Nama Narapidana: ')
  umur = input('Masukkan Umur Narapidana: ')
  asal = input('Masukkan Asal Narapidana: ')
  kejahatan = input('Masukkan Kejahatan yang dilakukan: ')
  lama_tahanan = input('Masukkan Lama Tahanan: ')

  narapidana_list.append({'nik': nik, 'nama': nama, 'umur': umur, 'asal': asal, 'kejahatan': kejahatan, 'lama_tahanan': lama_tahanan})
  print('Data Narapidana berhasil ditambahkan.')

# Menu 2
def delete_data(narapidana_list):
  print('\nHapus data Narapidana')
  delete = int(input('Masukkan data Narapidana ke berapa yang ingin dihapus: '))
  if delete < 1 or delete > len(narapidana_list):
    print('Data yang anda masukkan tidak valid.')
  else:
    del narapidana_list[delete-1]
    print(f'Data Narapidana ke-{delete} berhasil dihapus.')

# Menu 3
def display_data(narapidana_list):
  print('\nMenampilkan semua data Narapidana')
  print('Daftar Narapidana:')
  if len(narapidana_list) == 0:
    print("Data Narapidana kosong.")
    return
  for i in range(len(narapidana_list)):
      napi = narapidana_list[i]
      print(f"{i+1}. - NIK: {napi['nik']}\n   - NAMA: {napi['nama']}\n   - UMUR: {napi['umur']}\n   - ASAL: {napi['asal']}\n   - Jenis Kejahatan: {napi['kejahatan']}\n   - Lama Tahanan: {napi['lama_tahanan']}")

# Menu 4
def sort_nama(narapidana_list):
  if len(narapidana_list) <= 1:
    return narapidana_list
  mid = len(narapidana_list) // 2
  left_half = narapidana_list[:mid]
  right_half = narapidana_list[mid:]
  left_half = sort_nama(left_half)
  right_half = sort_nama(right_half)
  return merge_nama(left_half, right_half)

def merge_nama(left, right):
  hasil = []
  left_index, right_index = 0, 0
  while left_index < len(left) and right_index < len(right):
    if left[left_index]['nama'] < right[right_index]['nama']:
      hasil.append(left[left_index])
      left_index += 1
    else:
      hasil.append(right[right_index])
      right_index += 1
  while left_index < len(left):
    hasil.append(left[left_index])
    left_index += 1
  while right_index < len(right):
    hasil.append(right[right_index])
    right_index += 1
  return hasil

def sort_lamaTahanan(narapidana_list):
  if len(narapidana_list) <= 1:
    return narapidana_list
  mid = len(narapidana_list) // 2
  left_half = narapidana_list[:mid]
  right_half = narapidana_list[mid:]
  left_half = sort_lamaTahanan(left_half)
  right_half = sort_lamaTahanan(right_half)
  return merge_lamaTahanan(left_half, right_half)

def merge_lamaTahanan(left, right):
  hasil = []
  left_index, right_index = 0, 0
  while left_index < len(left) and right_index < len(right):
    if left[left_index]['lama_tahanan'] < right[right_index]['lama_tahanan']:
      hasil.append(left[left_index])
      left_index += 1
    else:
      hasil.append(right[right_index])
      right_index += 1
  while left_index < len(left):
    hasil.append(left[left_index])
    left_index += 1
  while right_index < len(right):
    hasil.append(right[right_index])
    right_index += 1
  return hasil

def sort_umur(narapidana_list):
  if len(narapidana_list) <= 1:
    return narapidana_list
  mid = len(narapidana_list) // 2
  left_half = narapidana_list[:mid]
  right_half = narapidana_list[mid:]
  left_half = sort_umur(left_half)
  right_half = sort_umur(right_half)
  return merge_umur(left_half, right_half)

def merge_umur(left, right):
  hasil = []
  left_index, right_index = 0, 0
  while left_index < len(left) and right_index < len(right):
    if left[left_index]['umur'] < right[right_index]['umur']:
      hasil.append(left[left_index])
      left_index += 1
    else:
      hasil.append(right[right_index])
      right_index += 1
  while left_index < len(left):
    hasil.append(left[left_index])
    left_index += 1
  while right_index < len(right):
    hasil.append(right[right_index])
    right_index += 1
  return hasil

# Menu 5
def searching_data(narapidana_list):
  print('\nCari data Narapidana')
  nik = input('Masukkan NIK narapidana yang akan dicari: ')
  for napi in narapidana_list:
    if napi['nik'] == nik:
      print(f'Data Narapidana dengan Nik-{nik} ditemukan:')
      print(f"   - NIK: {napi['nik']}\n   - NAMA: {napi['nama']}\n   - UMUR: {napi['umur']}\n   - ASAL: {napi['asal']}\n   - Jenis Kejahatan: {napi['kejahatan']}\n   - Lama Tahanan: {napi['lama_tahanan']}")
      return
  print(f'Data Narapidana dengan Nik-{nik} tidak ditemukan.')

# Menu 6
def update_data(narapidana_list):
  print('\nUpdate data Narapidana')
  update = int(input('Masukkan data Narapidana ke berapa yang ingin diupdate: '))
  if update < 1 or update > len(narapidana_list):
    print('Data yang anda masukkan tidak valid.')
  else:
    nik = input('Masukkan NIK Narapidana (Update): ')
    nama = input('Masukkan Nama Narapidana (Update): ')
    umur = input('Masukkan Umur Narapidana (Update): ')
    asal = input('Masukkan Asal Narapidana (Update): ')
    kejahatan = input('Masukkan Kejahatan yang dilakukan (Update): ')
    lama_tahanan = input('Masukkan Lama Tahanan (Update): ')

    narapidana_list[update-1] = {'nik': nik, 'nama': nama, 'umur': umur, 'asal': asal, 'kejahatan': kejahatan, 'lama_tahanan': lama_tahanan}
    print('Data Narapidana berhasil diupdate')
    print('Silahkan pilih menu 3 untuk menampilkan data yang telah diupdate.')


narapidana_list = []
print('\n\t--- PROGRAM INFORMASI NARAPIDANA ---')

while True:
    print('\nPilihan Menu:')
    print('1. Tambahkan Data Narapidana')
    print('2. Hapus Data Narapidana')
    print('3. Tampilkan Semua Data Narapidana')
    print('4. Urutkan Data Narapidana')
    print('5. Cari Data Narapidana')
    print('6. Update Data Narapidana')
    print('7. Exit/Keluar')
    pilihan = input('Masukkan pilihan anda (1-7): ')

    if pilihan == '1':
      insert_data(narapidana_list)
    elif pilihan == '2':
      delete_data(narapidana_list)
    elif pilihan == '3':
      display_data(narapidana_list)
    elif pilihan == '4':
      print('\nUrutkan data Narapidana berdasarkan: ')
      print('1. Nama')
      print('2. Lama Tahanan')
      print('3. Umur')
      print('4. Exit')
      pilih = input('Masukkan pilihan anda (1-3): ')
      if pilih == '1' :
        if len(narapidana_list) > 1:
          narapidana_list = sort_nama(narapidana_list)
          print('Data Narapidana berhasil diurutkan berdasarkan nama')
          print('Silahkan pilih menu 3 untuk menampilkan hasil.')
        else:
          print('Data Narapidana gagal diurutkan.')
      elif pilih == '2' :
        if len(narapidana_list) > 1:
          narapidana_list = sort_lamaTahanan(narapidana_list)
          print('Data Narapidana berhasil diurutkan berdasarkan lama tahanan')
          print('Silahkan pilih menu 3 untuk menampilkan hasil.')
        else:
          print('Data Narapidana gagal diurutkan.')
      elif pilih == '3' :
        if len(narapidana_list) > 1:
          narapidana_list = sort_umur(narapidana_list)
          print('Data Narapidana berhasil diurutkan berdasarkan umur')
          print('Silahkan pilih menu 3 untuk menampilkan hasil.')
        else:
          print('Data Narapidana gagal diurutkan.')
      elif pilih == '4':
        continue
      else :
        print('Pilihan tidak valid.')
    elif pilihan == '5':
      searching_data(narapidana_list)
    elif pilihan == '6':
      update_data(narapidana_list)
    elif pilihan == '7':
      berhenti = input('\nApakah Anda yakin ingin keluar dari program? (ya/tidak): ')
      if berhenti.lower() == 'ya':
        print('\nProgram Selesai.')
        print('--- TERIMA KASIH ---')
        break
      elif berhenti.lower() == 'tidak':
        narapidana_list = []
      else:
        print('Pilihan tidak valid. Silahkan ketik ya/tidak.')
    else:
      print('\nPilihan menu tidak valid.')
```
